```python
import os

# –ü—É—Ç–∞—ö–∞ –¥–æ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—ò—É–º–∞ —Å–∞ —Ç—Ä–µ–Ω–∏–Ω–≥ —Å–ª–∏–∫–∞–º–∞ –∏ –ª–∞–±–µ–ª–∞–º–∞
train_images_dir = '/content/cigarette-4/train/images/'
train_labels_dir = '/content/cigarette-4/train/labels/'

# –ö—Ä–µ–∏—Ä–∞—ò data.yaml —Ñ–∞—ò–ª –∑–∞ YOLOv8
data_yaml_content = f"""
train: {train_images_dir}
val: {train_images_dir}

nc: 1
names: ['cig_butt']
"""

data_yaml_path = '/content/data.yaml'

with open(data_yaml_path, 'w') as file:
    file.write(data_yaml_content)
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(data=data_yaml_path, epochs=10, imgsz=640)
```

```python
import cv2
import os

unlabeled_images_dir = '/content/drive/MyDrive/cig_butts/cig_butts/train/images'
output_labels_dir = '/content/drive/MyDrive/cig_butts/cig_butts/trained/generated_labels'

if not os.path.exists(output_labels_dir):
    os.makedirs(output_labels_dir)

def generate_labels(model, image_path, output_label_path):
    results = model(image_path)
    with open(output_label_path, 'w') as file:
        for result in results[0].boxes:
            x1, y1, x2, y2 = result.xyxy[0].int().tolist()
            file.write(f'0 {x1} {y1} {x2} {y2}\n')

for filename in os.listdir(unlabeled_images_dir):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        image_path = os.path.join(unlabeled_images_dir, filename)
        output_label_path = os.path.join(output_labels_dir, filename.replace('.jpg', '.txt').replace('.png', '.txt'))
        generate_labels(model, image_path, output_label_path)
        print(f'Generated label for {filename}')
```

```python
import cv2
import os
import random
import matplotlib.pyplot as plt

def visualize_image_with_labels(image_path, label_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    with open(label_path, 'r') as file:
        lines = file.readlines()

    for line in lines:
        label = line.strip().split()
        class_id, x1, y1, x2, y2 = map(float, label[1:])  # –ü—Ä–æ–º–µ–Ω–∞ —É float –∏ –ø—Ä–æ–º–µ–Ω–∞ —Ñ–æ—Ä–º–∞—Ç–∞
        h, w, _ = image.shape
        x1 = int(x1 * w)
        y1 = int(y1 * h)
        x2 = int(x2 * w)
        y2 = int(y2 * h)
        cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)

    return image

input_dir = '/content/drive/MyDrive/cig_butts/cig_butts/train/images'
labels_dir = '/content/drive/MyDrive/cig_butts/cig_butts/trained/generated_labels'

all_files = [f for f in os.listdir(input_dir) if f.endswith('.jpg') or f.endswith('.png')]

random_files = random.sample(all_files, 5)

for filename in random_files:
    image_path = os.path.join(input_dir, filename)
    label_path = os.path.join(labels_dir, filename.replace('.jpg', '.txt').replace('.png', '.txt'))
    if os.path.exists(label_path):
        print(f'Visualizing {filename}')
        image_with_labels = visualize_image_with_labels(image_path, label_path)
        plt.figure(figsize=(10, 10))
        plt.imshow(image_with_labels)
        plt.axis('off')
        plt.show()
        plt.close()
    else:
        print(f'Label not found for {filename}')
```
